```python
import mujoco as mj
import mujoco.viewer
import mediapy as media


model = mj.MjModel.from_xml_path('MJ_Leg_Smaller.xml')
data = mj.MjData(model)

# viewer = mujoco.viewer.launch(model, data)
```

Testing Camera Images

```python
# Create MuJoCo renderer
renderer = mj.Renderer(model)

# Show image of XML file for reference
mj.mj_forward(model, data)
renderer.update_scene(data, camera='Angle2')
media.show_image(renderer.render())
renderer.close() # Needed to prevent crashing?
```
