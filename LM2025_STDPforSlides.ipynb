```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(1000)
pre = np.zeros(1000)
post = np.zeros(1000)


pre[::300] = 1
post[510] = 1

plt.figure(figsize=[15, 5])
plt.plot(t/2, pre, linewidth=4, label='Presynaptic Neuron')
plt.plot(t/2, post, linewidth=4, label='Postsynaptic Neuron')
plt.ylabel('Spike', size=20)
plt.yticks(ticks=[])
plt.xlabel('Time (ms)', size=20)
plt.legend(loc='upper right', fontsize=15)
plt.show()


# color='darkorange',
# color='limegreen',
```
